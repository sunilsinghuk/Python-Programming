## **1. Python**
Python is a **beginner-friendly programming language** known for its **simplicity and readability**. Let’s explore some fundamental concepts to get started with Python programming.

### **1.1 Python Syntax**
**Syntax** refers to the **set of rules** that define how a Python program is written. Unlike some other programming languages, **Python relies heavily on indentation (spacing)** to define blocks of code, such as loops, functions, and conditions.

### **Key Syntax Points:**
* **Case Sensitivity:** Python is case-sensitive (Variable and variable are different).
* **Indentation:** Proper indentation is essential. Indentation is used to define blocks of code.

**Example**

```python
# A simple Python program
print("Hello, World!")  # This will print Hello, World!
```

**In Python, indentation is crucial. For example, if we have an if statement:**

```python
if 5 > 3:
    print("5 is greater than 3")
```

Notice that the code inside the if block is indented.

### **1.2 Variables in Python**
Variables are **containers for storing data**. In Python, you **don’t need to declare** the type of the variable (like in some other languages). You **just assign a value**, and Python will automatically recognize its type.

**Example:**

```python
# Creating a variable
name = "Alice"  # A string
age = 25        # An integer
height = 5.6    # A float

# Printing variables
print(name)  # Output: Alice
print(age)   # Output: 25
print(height)  # Output: 5.6
```

*   **Variables can store different types of data**, like text (strings), numbers (integers, floats), etc.
*   **Variable names** can contain letters, digits, and underscores, but cannot start with a digit.

**Example:**

```python
x = 10
y = 3.14
first_name = "John"
is_student = True
```

### **1.3 Data Types in Python**
In Python, **every value has a data type**. The most **common data types are:**

### **A. Numeric Data Types:**
* int: Whole numbers (e.g., 10, 0, -7)
* float: Numbers with decimals (e.g., 3.14, -0.1)
* complex: Complex numbers (e.g., 2 + 3j)

### **B. String Data Type:**
* str: Represents text (e.g., "hello", "Python")

### **C. Boolean Data Type:**
* bool: Represents True or False

### **D. List, Tuple, and Dictionary (Collections):**
* list: A collection of ordered, changeable items. (e.g., [1, 2, 3])
* tuple: A collection of ordered, unchangeable items. (e.g., (1, 2, 3))
* dict: A collection of key-value pairs. (e.g., {"name": "Alice", "age": 25})

**Example:**

```python
# Different data types
x = 10  # Integer
y = 3.14  # Float
name = "Alice"  # String
is_active = True  # Boolean
```

### **1.4 Type Checking and Type Conversion**
To check the type of a variable, **use the type() function**.

**Example:**

```python
x = 10
print(type(x))  # Output: <class 'int'>

y = "Hello"
print(type(y))  # Output: <class 'str'>
```

**To convert between data types,** use type conversion functions such as int(), float(), and str().

```python
age = "25"  # String
age = int(age)  # Convert to integer
print(type(age))  # Output: <class 'int'>
```

### **2.1 Arithmetic Operators**
Arithmetic operators perform mathematical operations.
image.png

[**To Check and Practice more examples of Operator, Click Here**](https://drive.google.com/file/d/1dhaK6Ol5-v0nZADG6K4tUruDZRYwC6vU/view?usp=sharing)

**Example:**

```python
a = 10
b = 3

print(a + b)  # Output: 13
print(a - b)  # Output: 7
print(a * b)  # Output: 30
print(a / b)  # Output: 3.33
print(a % b)  # Output: 1
```

### **2.2 Comparison Operators**
Comparison operators **compare two values and return a boolean** (True or False).
image.png

**Example:**

```python
x = 10
y = 5

print(x == y)  # Output: False
print(x > y)   # Output: True
```

### **2.3 Assignment Operators**
Assignment operators are **used to assign values to variables**.

image.png

**Example:**

```python
x = 10
x += 5  # Equivalent to x = x + 5
print(x)  # Output: 15
```

### **2.4 Logical Operators**
Logical operators are **used to combine conditional statements**.
image.png

**Example:**

```python
x = 5
y = 10

print(x > 3 and y < 15)  # Output: True
print(x > 3 or y < 5)    # Output: True
print(not(x > 3))        # Output: False
```

### **2.5 Identity Operators**
Identity operators **check if two variables refer to the same object**.

```python
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)   # Output: True (same object)
print(x is z)   # Output: False (different objects)
```
